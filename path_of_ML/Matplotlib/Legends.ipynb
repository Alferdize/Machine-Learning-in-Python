```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,6,3]
x2 = [1,2,3,4,5]
y2 = [2,4,6,8,1]
```

```python

plt.plot(x,y,label="First Line")
plt.plot(x2,y2,label="Second Line")
plt.xlabel('plot Numbers')
plt.ylabel("Important Value")
plt.title("Interesting Graph\nCheck it out")
plt.legend()
plt.show()
```
